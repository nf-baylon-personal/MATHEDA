# Normal Distribution

```python
cmean = cdf['Current'].mean()
cstd = cdf['Current'].std(ddof=1)

x_min = cdf['Current'].min()
x_max = cdf['Current'].max()

x = np.linspace(x_min,x_max,100)
p = stats.norm.pdf(x, cmean, cstd)

plt.plot(x, p,
         color = "#6dc4ff",
         linewidth = 2)


plt.title('Normal Distribution of Current Values',
          font = 'Courier New', fontsize = 12)

plt.xlabel('Current(mA)')
plt.ylabel('Probability Density')
plt.show()
```

```python
plt.hist(cdf['Current'], bins = 5,
         density=True,
         alpha = 0.8,
         color = '#fce300',
         edgecolor = '#000000')

plt.title('Normal Distribution of Current Values',
          font = 'Courier New', fontsize = 12)
plt.xlabel('Current(mA)')
plt.ylabel('Probability Density')

plt.plot(x, p,
         color = "#6dc4ff",
         linewidth = 2)

plt.show()
```

# Standard Normal DIstribution N(0,1)

```python
zmean = cdf['Z-score'].mean()
zstd = cdf['Z-score'].std(ddof=1)

z_min = cdf['Z-score'].min()
z_max = cdf['Z-score'].max()

xz = np.linspace(z_min,z_max,100)
pz = stats.norm.pdf(xz, zmean, zstd)

plt.plot(xz, pz,
         color = "#008533",
         linewidth = 2)


plt.title('Normal Distribution of Current Values N(0,1)',
          font = 'Courier New', fontsize = 12)

plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.show()
```
