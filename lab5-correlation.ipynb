```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv(r'dataset\current-power.csv', delimiter=",")
df
```

```python
df.describe()
```

```python
df.cov()
```

```python
x = df['Current']
y = df['Power']

plt.figure()
plt.scatter(x, y)
plt.title('Device Current vs Power Response')
plt.xlabel('Current (Ohms)')
plt.ylabel('Power (Watts)')
plt.show()
```

```python
df.corr()
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
line = slope * x + intercept

plt.figure(figsize=(7,4))
plt.scatter(x, y, color = '#000000', alpha = 0.6, label = 'Datapoints')
plt.plot(x, line, color = '#fce300', label = f'Regression line (r = {r_value:.3f})')
plt.title('Device Current vs Power Response')
plt.xlabel('Current (Ohms)')
plt.ylabel('Power (Watts)')
plt.legend()
plt.show()
```

## Air Quality

```python
air = pd.read_csv(r'dataset\air-quality.csv', delimiter=",")
air
```

### NOx-Humidity

```python
x = air['NOx']
y = air['Relative_Humidity']

slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
line = slope * x + intercept

plt.figure(figsize=(7,4))
plt.scatter(x, y, color = '#000000', alpha = 0.6, label = 'Datapoints')
plt.plot(x, line, color = '#fce300', label = f'Regression line (r = {r_value:.3f})')
plt.title('Nitrogen Oxide vs Humidity')
plt.xlabel('Nitrogen Oxide (ug/cu.m)')
plt.ylabel('Relative Humidity (%)')
plt.legend()
plt.show()
```

### NO2-Humidity

```python
x = air['NO2']
y = air['Relative_Humidity']

slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
line = slope * x + intercept

plt.figure(figsize=(7,4))
plt.scatter(x, y, color = '#000000', alpha = 0.6, label = 'Datapoints')
plt.plot(x, line, color = '#fce300', label = f'Regression line (r = {r_value:.3f})')
plt.title('Nitrogen Dioxide vs Humidity')
plt.xlabel('Nitrogen Dioxide (ug/cu.m)')
plt.ylabel('Relative Humidity (%)')
plt.legend()
plt.show()
```
